```python
exchange_rates = {
    "AA": 1,
    "AB": 1.45,
    "AC": 0.52,
    "AD": 0.72,
    "BA": 0.7,
    "BB": 1,
    "BC": 0.31,
    "BD": 0.48,
    "CA": 1.95,
    "CB": 3.1,
    "CC": 1,
    "CD": 1.49,
    "DA": 1.34,
    "DB": 1.98,
    "DC": 0.64,
    "DD": 1
}

keys = list(exchange_rates.keys())

max_product = 0           
best_sequence = []        
valid_sequences = []      

def find_sequences(current_seq):
    global max_product, best_sequence
    if len(current_seq) == 5:

        if current_seq[-1][1] == 'D':
            product = 1
            for key in current_seq:
                product *= exchange_rates[key]
            valid_sequences.append((current_seq.copy(), product))
            if product > max_product:
                max_product = product
                best_sequence = current_seq.copy()
        return

    last_letter = current_seq[-1][1]
    for k in keys:
        if k[0] == last_letter:
            current_seq.append(k)
            find_sequences(current_seq)
            current_seq.pop()

for k in keys:
    if k[0] == 'D':
        find_sequences([k])

print("All valid sequences and their product:")
for seq, prod in valid_sequences:
    print(seq, "=>", prod)

print("\nBest sequence:", best_sequence)
print("Max product:", max_product)
```
